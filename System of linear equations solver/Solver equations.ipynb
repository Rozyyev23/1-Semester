```python
import math

def sls(G, J):
    n = len(G)
    A = G.copy()
    b = J.copy()
    for i in range(n):
        
        sc = 1.0 / A[i][i]
        for j in range(n): 
            A[i][j] *= sc
        b[i] *= sc

        for k in range(n): 
            if k == i:
                continue
            sc = A[k][i]
            for j in range(n):
                A[k][j] -= sc * A[i][j]
            b[k] -= sc * b[i]

    ls = b
    
    return ls

def quadratic(kef):
    a, b, c = kef
    D = b ** 2 - 4 * a * c
    print(D)
    if D < 0:
        return ['Корней нет']
    elif D == 0:
        x1 = -b / 2 * a
        return [x1]
    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2 * a)
        x2 = (-b - math.sqrt(D)) / (2 * a)
        
        return [x1, x2]

def cubic(kef):
    a, b, c, d = kef 

    f = ((3.0 * c / a) - ((b ** 2.0) / (a ** 2.0))) / 3.0
    g = (((2.0 * (b ** 3.0)) / (a ** 3.0)) - ((9.0 * b * c) / (a **2.0)) + (27.0 * d / a)) /27.0
    h = ((g ** 2.0) / 4.0 + (f ** 3.0) / 27.0)

    if f == 0 and g == 0 and h == 0:            
        if (d / a) >= 0:
            x = (d / (1.0 * a)) ** (1 / 3.0) * -1
        else:
            x = (-d / (1.0 * a)) ** (1 / 3.0)
        return [x, x, x]

    elif h <= 0:                               

        i = math.sqrt(((g ** 2.0) / 4.0) - h)
        j = i ** (1 / 3.0)
        k = math.acos(-(g / (2 * i)))
        L = j * -1
        M = math.cos(k / 3.0)
        N = math.sqrt(3) * math.sin(k / 3.0)
        P = (b / (3.0 * a)) * -1

        x1 = 2 * j * math.cos(k / 3.0) - (b / (3.0 * a))
        x2 = L * (M + N) + P
        x3 = L * (M - N) + P

        return [x1, x2, x3]

    elif h > 0:                                 
        R = -(g / 2.0) + math.sqrt(h) 
        if R >= 0:
            S = R ** (1 / 3.0)                  
        else:
            S = (-R) ** (1 / 3.0) * -1          
        T = -(g / 2.0) - math.sqrt(h)
        if T >= 0:
            U = (T ** (1 / 3.0))                
        else:
            U = ((-T) ** (1 / 3.0)) * -1

        x1 = (S + U) - (b / (3.0 * a))
        x2 = -(S + U) / 2 - (b / (3.0 * a)) + (S - U) * math.sqrt(3) * 0.5j
        x3 = -(S + U) / 2 - (b / (3.0 * a)) - (S - U) * math.sqrt(3) * 0.5j

        return [x1, x2, x3]
    

def quartic(kef):
    a, b, c = kef
    t = a**2
    D = b ** 2 - 4 * a * c
    print(D)
    if D == 0:
        x = -b / 2 * a
        x1 = x**(1/2)
        x2 = -(x)**(1/2)
        return [x1, x2]
    else:
        p = (-b + math.sqrt(D)) / (2 * a)
        n = (-b - math.sqrt(D)) / (2 * a)
    x1 = p**(1/2)
    x2 = -(p)**(1/2)
    x3 = n**(1/2)
    x4 = -(n)**(1/2)
    return [x1, x2, x3, x4]

def linears(es):
    b = []
    A = []
    for eq in es:
        A.append([])
        eq_split = eq.split('=')
        b.append(float(eq_split[-1].strip()))
        sign = 1
        for x in eq_split[0].split(' '):
            if x != '+' and x != '-' and len(x) != 0:
                if '*' in x:
                    A[-1].append(sign*float(x.split('*')[0]))
                else:
                    try:
                        b[-1] -= sign*float(x)
                    except ValueError:
                        if '-' in x:
                            A[-1].append(-1.0*sign)
                        else:
                            A[-1].append(1.0*sign)
            elif x == '+':
                sign = 1
                
            elif x == '-':
                sign = -1
                
            else:
                continue
                
    return A, b

def nonlinears(ess):
    eq_split = ess.split('=')
    d = [-1*float(eq_split[-1].strip())]
    kef = []

    sign = 1
    for em in eq_split[0].split(' '):
        if em != '+' and em != '-' and len(em) != 0:
            if '*' in em:
                try:
                    coefficient = em.split('*')[0]
                    if '/' in coefficient:
                        frac = coefficient.split('/')
                        coefficient = frac[0]/frac[1]
                        kef.append(sign*coefficient)
                    else:
                        kef.append(sign*float(coefficient))
                except ValueError:
                    continue
            else:
                try:
                    d.append(sign*float(em))
                except ValueError:
                    if '-' in em:
                        kef.append(-1.0*sign)
                    else:
                        kef.append(1.0*sign)
        elif em == '+':
            sign = 1

        elif em == '-':
            sign = -1

        else:
            continue
           
    kef.append(sum(d))
    return kef

def resp(line_sol, nonline_sol):
    resp1 = []
    for el in nonline_sol:
        mm = line_sol.copy()
        mm.append(el)
        resp1.append(mm)
    return resp1

def systema(filename): 
    nl = 'x^3 + y^3 + z^3 + x^2 + y^2 + z^2 + x + y +3*z + 5 = 10'
    
    
    linear = [] 
    nl = None
    A, b = [], []
    with open(filename, 'r', encoding='utf-8') as file:
        for i in file:
            if '^' not in i: 
                linear.append(i.strip()) 
            else:
                nl = i.strip() 
    
    G, J = linears(linear)
    ls = sls(G, J)  
    
    if '^2' in nl and '^3' in nl and '^4' not in nl: 
        ns = cubic(nonlinears(nl))
    elif '^2' in nl and '^3' not in nl and '^4' not in nl:
        ns = quadratic(nonlinears(nl))
    elif '^2' in nl and '^3' not in nl and '^4' in nl:
        ns = quartic(nonlinears(nl))
    resp1 = resp(ls, ns)
    
    return resp1
systema('./final')
```
