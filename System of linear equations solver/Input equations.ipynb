```python
number = int(input("Введите количество уравнений в системе\n"))
print("Правила написания систем:")
print(
    "1. В качестве переменных можно использовать буквы латинского алфавита, либо индексацию одной переменной."
)
print(
    '2. Возможно использовать такие операции как "+" - плюс, "-" - минус, "^" - возведение в степень. Как ими пользоваться будет написано ниже.'
)
print(
    '3. Операции плюс и минус допустимы для сложения и вычитания как переменых так и констант.'
)
print(
    '4. Операция возведения в степень возможна только для переменных. Нельзя делать переменную отрицательной!'
)
print(
    '5. Возможно написание перед переменными агрумент. В качестве аргумента может служить любое вещественное число. Между переменной и ее агрументов ставить знак "*" - обязательно.'
)
print("Пример правильного написания уравнения в системе: -23*x^2 -3*y + 12 = 0")
print(
    "Пример неправильного написания уравнение в системе: -23x^(-2) + (-3y) + 12 = 0"
)
sistema = []
for i in range(number):
    sistema.append(input(f"Введите {i + 1}-e уравнение:\n"))

with open("final", "w") as outfile:
    outfile.write("\n".join(sistema)) #Записывает вводимые данные в файл вида .txt,
```
